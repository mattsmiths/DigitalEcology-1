```python
phrase = "" # @param {type:"string"}
```

```python
umFinal = []
nFinal = []

phrase1 = 'um'
in1 = open('/content/biol470/week9/Lecture_5_recording_Feb11.txt','r')
out = in1.readlines()
in1.close()

umCount = 0
nCount = 0
for line in out:
  if phrase in line:
    umCount += 1
  if phrase1 in line:
    nCount += 1
print('Normalized sleep: -0.64223556')
print(umCount)
umFinal.append(umCount)
nFinal.append(nCount)

in1 = open('/content/biol470/week9/Lecture_3_recording_Jan28.txt','r')
out = in1.readlines()
in1.close()

umCount = 0
nCount = 0
for line in out:
  if phrase in line:
    umCount += 1
  if phrase1 in line:
    nCount += 1
print('Normalized sleep: 1.2344268')
print(umCount)
umFinal.append(umCount)
nFinal.append(nCount)

in1 = open('/content/biol470/week9/Activity_5_recording_Feb13.txt','r')
out = in1.readlines()
in1.close()

umCount = 0
nCount = 0
for line in out:
  if phrase in line:
    umCount += 1
  if phrase1 in line:
    nCount += 1
print('Normalized sleep: -1.26778969')
print(umCount)
umFinal.append(umCount)
nFinal.append(nCount)

in1 = open('/content/biol470/week9/Lecture 4_recording_Feb4.txt','r')
out = in1.readlines()
in1.close()

umCount = 0
nCount = 0
for line in out:
  if phrase in line:
    umCount += 1
  if phrase1 in line:
    nCount += 1

print('Normalized sleep: 0.67559845')
print(umCount)
umFinal.append(umCount)
nFinal.append(nCount)




from matplotlib import pyplot as plt
import numpy as np

plt.figure(figsize=(10,6))
plt.scatter([-0.64223556,1.2344268,  -1.26778969,  0.67559845],umFinal,s=60,color=(0.35,0.5,0.4,0.5),label=phrase)
vx = [-0.64223556,1.2344268,  -1.26778969,  0.67559845]
xx = np.argsort(vx)
vx1 = np.array(vx)[xx]
umFinal = np.array(umFinal)[xx]
plt.plot(vx1,umFinal,linewidth=1,color=(0.35,0.5,0.4,0.5))


plt.scatter([-0.64223556,1.2344268,  -1.26778969,  0.67559845],nFinal,s=60,color=(0.7,0.5,0.4,0.5),label=phrase1)
vx = [-0.64223556,1.2344268,  -1.26778969,  0.67559845]
xx = np.argsort(vx)
vx = np.array(vx)[xx]
nFinal = np.array(nFinal)[xx]
plt.plot(vx,nFinal,linewidth=2,color=(0.7,0.5,0.4,0.5))
plt.legend(fontsize=12)

plt.ylabel('Word "%s" count'%phrase,fontsize=12)
plt.xlabel('Normalized sleep',fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.plot([vx[0],vx[-1]],[umFinal[0],umFinal[-1]],color=(0.35,0.5,0.4,0.5))
plt.scatter([vx[0],vx[-1]],[umFinal[0],umFinal[-1]],s=60,color=(0.35,0.5,0.4,0.5),label=phrase)

plt.legend(fontsize=12)
r,p = scipy.stats.pearsonr(umFinal,vx)
x1 = np.nanmean([vx[0],vx[-1]])
y1 = np.nanmean([umFinal[0],umFinal[-1]])

ax1 = plt.axis()
dd = (ax1[3]-ax1[2])*0.1
plt.text(x1,y1,'r='+str(round(r,3)),fontsize=12)
plt.text(x1,y1-dd,'p='+str(round(p,3)),fontsize=12)


plt.figure(figsize=(5,5))
plt.plot([vx1[0],vx1[-1]],[nFinal[0],nFinal[-1]],color=(0.7,0.5,0.4,0.5))
plt.scatter([vx1[0],vx1[-1]],[nFinal[0],nFinal[-1]],s=60,color=(0.7,0.5,0.4,0.5),label=phrase1)

ax1 = plt.axis()
dd = (ax1[3]-ax1[2])*0.1
r,p = scipy.stats.pearsonr(nFinal,vx1)
x1 = np.nanmean([vx1[0],vx1[-1]])
y1 = np.nanmean([nFinal[0],nFinal[-1]])
plt.text(x1,y1,'r='+str(round(r,3)),fontsize=12)
plt.text(x1,y1-dd,'p='+str(round(p,3)),fontsize=12)

plt.legend(fontsize=12)

plt.ylabel('Word "%s" count'%phrase,fontsize=12)
plt.xlabel('Normalized sleep',fontsize=12)
plt.show()
```
